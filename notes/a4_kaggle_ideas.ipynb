```python
import numpy as np
import pandas as pd

data1 = pd.read_csv('a4_data_kaggle.csv')
data1
```

```python
countries = dict(list(data1.groupby('country')))
countries.keys()
```

```python
countries['Albania']
```

```python
for key in countries:
    countries[key].drop('country', axis = 1, inplace = True)
```

```python
for key in countries:
    countries[key] = countries[key].pivot_table(index = ['year','age'], columns = 'generation', values = 'suicides')
```

```python
missing_years = []

for year in range(2000,2017):
    print(year)
    for key in countries:
        try:
            countries[key].loc[year]
        except:
            missing_years.append((key, year))
```
